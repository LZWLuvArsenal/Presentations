```python
import requests
import json
import pandas as pd
```

```python
url = "https://api.data.gov.sg/v1/environment/relative-humidity?date=2020-01-01"
```

```python
res = requests.get(url)
```

```python
resjon = res.json()
print(resjon)
```

```python
with open('01Jan2020_Relative_Humidity.json', 'w') as write_file:
  json.dump(resjon, write_file)
```

```python
with open("01Jan2020_Relative_Humidity.json", "r") as read_file:
    data = json.load(read_file)
    
 # Print the type of data variable 
    print("Type:", type(data)) 
```

```python
# Print the data of dictionary 
# for reading nested data [0] represents 
# the index value of the list 

print(data['items'][0]) 
```

```python
metadata_humidity = (data['metadata'])
items_humidity = (data['items'])
```

```python
with open('01Jan2020_Relative_Humidity_metadata_humidity.json', 'w') as write_file:
  json.dump(metadata_humidity, write_file)
```

```python
with open('01Jan2020_Relative_Humidity_items_humidity.json', 'w') as write_file:
  json.dump(items_humidity, write_file)
```
